# Exploratory Data Analysis (EDA) on World Population Data (1970–2024) using Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\usero\Desktop\current_world_population.csv")
df
```

```python
# Get general information about the df
df.info()
```

```python
# Set display option to show floats with 2 decimal places
pd.set_option('display.float_format', lambda x : '%.2f' % x)
```

```python
# Generate descriptive statistics of the df's numeric columns
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Get the number of unique values in each column of the df
df.nunique()
```

```python
# Get top 10 entries with the largest population in 2024
df.sort_values(by="2024 Population", ascending = False).head(10)
```

```python
# Compute the correlation matrix for numeric columns
df.corr(numeric_only=True)
```

```python
# Display a heatmap of the correlation matrix for all numeric columns
sns.heatmap(df.corr(numeric_only=True), annot= True)
plt.rcParams['figure.figsize'] = (20, 8)
plt.show()
```

```python
# Identify continents with the fastest-growing populations
df.groupby('Continent').mean(numeric_only=True).sort_values(by = '2024 Population', ascending=False)
```

```python
# Show only rows where the continent is Europe
df[df['Continent'].str.contains('Europe')]
```

```python
# Display all column names
df.columns
```

```python
# Plot population growth trends by continent from 1970 to 2024
# df2 = df.groupby('Continent')[df.columns[5:13]].mean(numeric_only=True).sort_values(by = '2024 Population', ascending=False)
df2 = df.groupby('Continent')[
        ['1970 Population', '1980 Population', '1990 Population',
         '2000 Population', '2010 Population', '2015 Population',
         '2020 Population', '2022 Population', '2024 Population']
    ].mean(numeric_only=True).sort_values(by='2024 Population', ascending=False)

df2.transpose().plot()
```

```python
# Create a boxplot for all numeric columns
df.boxplot(figsize=(20,8))
```
